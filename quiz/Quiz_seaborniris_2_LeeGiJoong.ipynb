```python
# enviroment setting 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

# Quiz 1. make pivot table 
- iris_df 데이터 프레임을 활용하여서 종에 대한 피봇 테이블을 만들어 주세요!
- 아래와 같은 결과가 나오면 됩니다
- 값들은 평균값들이 나왔으면 좋겠습니다!

```python
# code
```

```python
pd.pivot_table(iris_df, index=['species'], aggfunc=np.mean)
```

# Quiz 2. make boxplot
- iris_df를 활용해서 boxplot을 그려주세요!
- x_values = sepal_length , y_values  = petal_width 
- 종으로 분류 해주세요!   

```python
plt.figure(figsize = (14,8))
# code 
```

```python
plt.figure(figsize = (14,8))
# code 
sns.boxplot(x = 'sepal_length' , y = 'petal_width', data=iris_df, hue='species')
plt.show()
```

#  Quiz 4. make pairplot
- iris_df를 활용하여서 pairplot을 그려주세요!
- x_values = sepal_length','sepal_width, y_values  = petal_width
- 종으로 분류 해주세요!    

```python
plt.figure(figsize = (14,8))

# code
```

```python
plt.figure(figsize = (14,8))

# code
sns.pairplot(x_vars=['sepal_length','sepal_width'], y_vars='petal_width', data=iris_df, hue='species')
```

#  Quiz 5. make plot 
- iris_df를 활용하여서 plot 그림을 그려주세요 ('species' 컬럼은 드랍하셔야 그림이 그려집니다)
- plot 그림을 그리는데 데이터 프레임의 모든 각각의 값은 자연상수의 제곱으로 보고 싶어요! (hint np.exp(data)

```python
# code
iris_df2 = np.exp(iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
iris_df2.plot(figsize = (16,10))
plt.show()
```
