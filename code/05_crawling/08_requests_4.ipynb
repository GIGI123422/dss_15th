###  HTML 데이터 파싱을 이용한 크롤링
- 네이버 연관검색어 >> 네이버 연관검색어 없어짐 >> 다음으로 대체
    - 1. 웹페이지 분석: URL 찾기
        - 개발자 모드 >> 네트워크 탭 >> all탭 >> 리스트의 맨위
    - 2. 요청 >> 응답: HTML(str)
    - 3. HTML(str) >> BeautifulSoup obj(css-selector) >> DF

```python
# bs보다 requests가 위에 있어야함
import requests
from bs4 import BeautifulSoup 
```

```python
# 1. 웹페이지 분석: URL
query='삼성전자'
# 네이버 연관검색어 이젠 없음
# url='https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&\
# query={}&oquery={}&tqi=UHqS7lp0JywssjLkuJVssssstCC-181913'.format(query, query)

url = 'https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q={}'.format(query)
print(url)
```

```python
# 2. request -> response : HTML(str)
response = requests.get(url)
response
```

```python
response.text
```

```python
# 3. HTML(str) >> BeautifulSoup Obj (css-selector) >> DF
dom = BeautifulSoup(response.content, features = "html.parser")
type(dom), type(response.text), type(response.content), type(response)
```

```python
# select: 여러개의 element 객체를 리스트로 가져옴
# select_one: 하나의 element 객체를 가져옴
```

```python
# 함수로 만들기
def daum_relational_key(query):
    url = 'https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q={}'.format(query)
    response = requests.get(url)
    dom = BeautifulSoup(response.text, "html.parser")
    elements = dom.select('span.wsn')
    return[element.text.strip() for element in elements]
```
