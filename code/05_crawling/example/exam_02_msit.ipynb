#### msit: 과학기술 정보 통신부
- https://msit.go.kr
- 보도자료 크롤링
- json으로 데이터 수집
- json 데이터 안에 html 코드가 들어있는 건에 대한 예제

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# 1. URL 가져오기
    # tip: query문에 value가 없는것들은 빼서 URL 길이를 보기좋게 줄이자
    # JQuery 그냥 날림
page = 1
url = "https://www.msit.go.kr/dynamic/article/_policycom3?\
&search=creday<=20201030-194500&sort=operday=desc&&idx=desc&below=false&pageout=true\
&aradon.result.format=json&aradon.page.pageNo={}&aradon.page.listNum=10&aradon.page.screenCount=5\
&aradon.referrer.href=https://www.msit.go.kr/web/msipContents/contents.do?mId=MTAxOQ==&_=1604056558337".format(page)
# 2. request, response 
response = requests.get(url)
# 3. 파싱
result = response.json()["result"]["nodes"]
result
```

```python
# DataFrame
columns=["action", "admin_p", "artid", "artsubject", "artcont"]
msit_df = pd.DataFrame(result)[columns]
msit_df.tail(2)
```

```python
msit_df["artcont"][0]
```

```python
html = msit_df["artcont"][0] # html 데이터만 선택
remove_words = ["\r", "\xa0", "\n", "□"]
content = BeautifulSoup(html, "html.parser").text # 태그의 값만 가져옴
for remove_word in remove_words:
    content = content.replace(remove_word, "")
content
```

```python
def html_to_text(html):
    remove_words = ["\r", "\xa0", "\n", "□"]
    content = BeautifulSoup(html, "html.parser").text # 태그의 값만 가져옴
    for remove_word in remove_words:
        content = content.replace(remove_word, "")
    return content
```

```python
html = msit_df["artcont"][1] # html 데이터만 선택
content = html_to_text(html)
content[:100]
```

```python
msit_df["artcont"] = msit_df["artcont"].apply(html_to_text)
msit_df.tail(2)
```
